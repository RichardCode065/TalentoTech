# ***Calculadora con funciones***

```python
import math

# --------------------- MENÚ ---------------------
def menuPrincipal():
    print("\nCalculadora Media")
    print("1. Sumar")
    print("2. Restar")
    print("3. Multiplicar")
    print("4. Dividir")
    print("5. Potenciar")
    print("6. Raíz cuadrada")
    print("7. Logaritmo (base 10)")
    print("8. Seno - sin")
    print("9. Coseno - cos")
    print("10. Tangente - tan")
    print("11. Salir")

# Validacion de Entradas

def ingresoUnNumero(mensaje="Ingrese un número: "):
    while True:
        try:
            return float(input(mensaje))
        except ValueError:
            print("Entrada inválida.")

def ingresoDosNumeros():
    numeroUno = ingresoUnNumero("Primer número: ")
    numeroDos = ingresoUnNumero("Segundo número: ")
    return numeroUno, numeroDos

# Funciones de la calculadora

def sumar():
    numeroUno, numeroDos = ingresoDosNumeros()
    print(f"Resultado: {numeroUno + numeroDos}")

def restar():
    numeroUno, numeroDos = ingresoDosNumeros()
    print(f"Resultado: {numeroUno - numeroDos}")

def multiplicar():
    numeroUno, numeroDos = ingresoDosNumeros()
    print(f"Resultado: {numeroUno * numeroDos}")

def dividir():
    numeroUno, numeroDos = ingresoDosNumeros()
    if numeroDos != 0:
        print(f"Resultado: {numeroUno / numeroDos}")
    else:
        print("Error: División por cero.")

def potenciar():
    numeroUno, numeroDos = ingresoDosNumeros()
    print(f"Resultado: {math.pow(numeroUno, numeroDos)}")

def raizCuadrada():
    numeroUno = ingresoUnNumero()
    if numeroUno >= 0:
        print(f"Resultado: {math.sqrt(numeroUno)}")
    else:
        print("Error: Raíz no definida para negativos.")

def logaritmo():
    numeroUno = ingresoUnNumero()
    if numeroUno > 0:
        print(f"Resultado: {math.log10(numeroUno)}")
    else:
        print("Error: Logaritmo solo definido para positivos.")

def seno():
    numeroUno = ingresoUnNumero("Ingrese el ángulo en grados: ")
    radianes = math.radians(numeroUno)
    print(f"sen({numeroUno}°) = {math.sin(radianes)}")

def coseno():
    numeroUno = ingresoUnNumero("Ingrese el ángulo en grados: ")
    radianes = math.radians(numeroUno)
    print(f"cos({numeroUno}°) = {math.cos(radianes)}")

def tangente():
    numeroUno = ingresoUnNumero("Ingrese el ángulo en grados: ")
    if numeroUno % 180 == 90:
        print("Error: La tangente no está definida para 90° + n*180°.")
    else:
        radianes = math.radians(numeroUno)
        print(f"tan({numeroUno}°) = {math.tan(radianes)}")

# Llamado de funciones

def ejecutarCalculadora(opcion):
    match opcion:
        case "1": sumar()
        case "2": restar()
        case "3": multiplicar()
        case "4": dividir()
        case "5": potenciar()
        case "6": raizCuadrada()
        case "7": logaritmo()
        case "8": seno()
        case "9": coseno()
        case "10": tangente()
        case "11":
            print("Gracias, Vuelva pronto!")
            return False
        case _:
            print("Opción inválida. Ingrese una opción válida")
    return True

# Funcion Principal

def Calculadora():
    continuar = True
    while continuar:
        menuPrincipal()
        opcion = input("Seleccione una opción: ")
        continuar = ejecutarCalculadora(opcion)

# Ejecución

Calculadora()
```
