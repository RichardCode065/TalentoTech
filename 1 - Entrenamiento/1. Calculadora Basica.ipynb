```python
# Funcion encargada del menu
# Creación del menu usando funciones

def menuPrincipal():
    print("\nCalculadora Media")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Salir")

# Ciclo para mantener activa la calculadora hasta que el
# Usuario desee terminar la ejecución de los cálculos matemáticos

while True:
    # Invocación de la función menuPrincipal()
    menuPrincipal()
    elegirOpcionMenu = input("Seleccione una opción: ")

    # Salir del programa
    if elegirOpcionMenu == '5':
        print("Gracias por usar esta calculadora!")
        break

    # Validación de opciones válidas
    if elegirOpcionMenu in ["1", "2", "3", "4"]:
        try:
            numeroUno = float(input("Ingrese el primer número: "))
            numeroDos = float(input("Ingrese el segundo número: "))

            # S U M A
            if elegirOpcionMenu == '1':
                print(f"Resultado: {numeroUno + numeroDos}")

            # R E S T A
            elif elegirOpcionMenu == '2':
                print(f"Resultado: {numeroUno - numeroDos}")

            # M U L T I P L I C A C I Ó N
            elif elegirOpcionMenu == '3':
                print(f"Resultado: {numeroUno * numeroDos}")

            # D I V I S I Ó N
            elif elegirOpcionMenu == '4':
                if numeroDos != 0:
                    print(f"Resultado: {numeroUno / numeroDos}")
                else:
                    print("Error: No se puede dividir entre 0.")

        except ValueError:
            print("Error: Ingrese solo números válidos.")
    else:
        print("Opción inválida. Intente de nuevo.")
```
