### ***Invento de programa "Listado de Notas"***
## ***F u n c i o n a l i d a d e s***
* Nota más alta
* Nota más baja
* Mejor nota
* Mensajes de acuerdo a la nota del estudiante
* Mejor estudiante
* Peor estudiante


```python
# Versión 1

# Lista de estudiantes (cada uno es un diccionario)
estudiantesLista = [
    {"nombreEstudiante": "Laura", "notaEstudiante": 4.5},
    {"nombreEstudiante": "Santiago", "notaEstudiante": 3.2},
    {"nombreEstudiante": "Valeria", "notaEstudiante": 5.0},
    {"nombreEstudiante": "Mateo", "notaEstudiante": 2.8},
    {"nombreEstudiante": "Camila", "notaEstudiante": 4.0},
    {"nombreEstudiante": "Andrés", "notaEstudiante": 3.6},
    {"nombreEstudiante": "Sofía", "notaEstudiante": 4.8},
    {"nombreEstudiante": "Juan", "notaEstudiante": 2.5},
    {"nombreEstudiante": "Isabela", "notaEstudiante": 4.2},
    {"nombreEstudiante": "Tomás", "notaEstudiante": 3.1},
    {"nombreEstudiante": "Daniela", "notaEstudiante": 4.9},
    {"nombreEstudiante": "Felipe", "notaEstudiante": 2.9},
    {"nombreEstudiante": "Mariana", "notaEstudiante": 3.5},
    {"nombreEstudiante": "Samuel", "notaEstudiante": 4.3},
    {"nombreEstudiante": "Lucía", "notaEstudiante": 3.0}
]

# Tupla para comentarios (según rendimiento)
comentarios = (
    "¡Bien hecho!",
    "¡No te rindas! ",
    "¡Vámos tú puedes! "
)


# Variables de cálculo
sumaNotas = 0
notaMasAlta = 0
notaMasBaja = 5.0
estudianteSuperior = ""
estudianteBajo = ""
totalEstudiantes = len(estudiantesLista)

for studentF in estudiantesLista:
    nota = studentF["notaEstudiante"]
    sumaNotas += nota

    if nota > notaMasAlta:
        notaMasAlta = nota
        estudianteSuperior = studentF["nombreEstudiante"]

    if nota < notaMasBaja:
        notaMasBaja = nota
        estudianteBajo = studentF["nombreEstudiante"]

promedio = sumaNotas / totalEstudiantes

# Mostrar resultados
print("Listado de estudiantes y notas:\n")
for studentF in estudiantesLista:
    print(f"{studentF['nombreEstudiante']}: {studentF['notaEstudiante']}")

print("\nResultados generales:")
print(f"Promedio del grupo: {promedio:.2f}")
print(f"Nota más alta: {notaMasAlta} ({estudianteSuperior})")
print(f"Nota más baja: {notaMasBaja} ({estudianteBajo})")

# Retroalimentación
print("\nRetroalimentación:")
for studentF in estudiantesLista:
    nota = studentF["notaEstudiante"]
    if nota >= 4.5:
        print(f"{studentF['nombreEstudiante']}: {comentarios[0]}")
    elif nota >= 3.0:
        print(f"{studentF['nombreEstudiante']}: {comentarios[1]}")
    else:
        print(f"{studentF['nombreEstudiante']}: {comentarios[2]}")
```

```python
#Version 2

# Tupla para comentarios (según rendimiento)
comentarios = (
    "¡Bien hecho!",
    "¡No te rindas! ",
    "¡Vámos tú puedes! "
)

# Lista para almacenar estudiantes
estudiantesLista = []

# 1. Solicitar número de estudiantes
totalEstudiantes = int(input("¿Cuántos estudiantes vas a registrar?: "))

# 2. Ingreso de datos por usuario
for i in range(totalEstudiantes):
    print(f"\nEstudiante {i+1}:")
    nombreEstudiante = input("Nombre: ")
    while True:
        try:
            notaEstudiante = float(input("Nota (0.0 a 5.0): "))
            if 0.0 <= notaEstudiante <= 5.0:
                break
            else:
                print("La nota debe estar entre 0.0 y 5.0.")
        except ValueError:
            print("Ingresa un número válido.")
    estudiantesLista.append({"nombre": nombreEstudiante, "nota": notaEstudiante})

# Conjunto de nombres únicos
nombreUnico = {estudiantesCicloF["nombre"] for estudiantesCicloF in estudiantesLista}

# Cálculos
sumaNotas = 0
notaMasAlta = 0
notaMasBaja = 5.0
estudianteSuperior = ""
estudianteBajo = ""

for estudiantesCicloF in estudiantesLista:
    notaEstudiante = estudiantesCicloF["nota"]
    sumaNotas += notaEstudiante
    if notaEstudiante > notaMasAlta:
        notaMasAlta = notaEstudiante
        estudianteSuperior = estudiantesCicloF["nombre"]
    if notaEstudiante < notaMasBaja:
        notaMasBaja = notaEstudiante
        estudianteBajo = estudiantesCicloF["nombre"]

promedio = sumaNotas / len(estudiantesLista)

# Resultados
print("\nListado de estudiantes y notas:")
for estudiantesCicloF in estudiantesLista:
    print(f"{estudiantesCicloF['nombre']}: {estudiantesCicloF['nota']}")

print("\nResultados generales:")
print(f"Promedio del grupo: {promedio:.2f}")
print(f"Nota más alta: {notaMasAlta} ({estudianteSuperior})")
print(f"Nota más baja: {notaMasBaja} ({estudianteBajo})")

print("\nRetroalimentación:")
for estudiantesCicloF in estudiantesLista:
    if estudiantesCicloF["nota"] >= 4.5:
        print(f"{estudiantesCicloF['nombre']}: {comentarios[0]}")
    elif estudiantesCicloF["nota"] >= 3.0:
        print(f"{estudiantesCicloF['nombre']}: {comentarios[1]}")
    else:
        print(f"{estudiantesCicloF['nombre']}: {comentarios[2]}")
```
