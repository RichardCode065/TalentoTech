###***T u p l a***

```python
# Tupla: colección inmutable (no se puede modificar)
frutas = ("manzana", "banano", "toronja", "cereza", "kiwi", "naranja", "frambuesa", "piña", "mango", "sandía")

# Imprimir un elemento específico
print("Fruta en la posición 3:", frutas[3])

# Recorrer todos los elementos de la tupla
for fruta in frutas:
    print(fruta)

```

###***L i s t a***

```python
# Lista: colección ordenada y mutable

nombres = ["Marcela", "Luisa", "Pedro", "Lucía", "Carlos", "María", "José", "Camila", "Estefania", "Elena"]

# Agregar un nuevo nombre al final de la lista
nombres.append("Valentina")

# Modificar un elemento (reemplazar a "Pedro" por "Andrés")
nombres[2] = "Andrés"

# Ordenar alfabéticamente la lista
nombres.sort()

# Imprimir la lista ordenada
print("Lista ordenada:", nombres)

# Recorrer todos los nombres
for nombre in nombres:
    print(nombre)
```

###***C o n j u n t o***

```python
# Set: colección sin orden y sin elementos repetidos
color = {"rojo", "verde", "azul", "amarillo", "negro", "blanco", "gris", "naranja", "morado", "rosa"}

# Agregar un color al conjunto
color.add("turquesa")

# Eliminar un color del conjunto
color.discard("gris")

# Verificar si un color está presente
print("¿'rojo' está en colores?", "rojo" in color)

# Recorrer todos los elementos del conjunto
for color in color:
    print(color)
```

###***D i c c i o n a r i o***

```python
# Diccionario: colección de pares clave:valor
estudiantes = {
    "Marcela": 20,
    "Luisa": 21,
    "Pedro": 19,
    "Lucía": 22,
    "Carlos": 18,
    "María": 23,
    "José": 20,
    "Camila": 21,
    "Estefania": 22,
    "Elena": 19
}

# Agregar un nuevo estudiante
estudiantes["Valentina"] = 20

# Modificar la edad de un estudiante
estudiantes["Carlos"] = 19

# Recorrer todos los pares clave:valor
for nombre, edad in estudiantes.items():
    print(f"{nombre} tiene {edad} años")
```
